```python
with open('dane_numeryczne.csv', 'r') as plik:
  plik.readline();
  odczytane_dane = plik.readlines();
  dane = [];
  for el in odczytane_dane:
    oczyszczony_wiersz = el.strip().split(",");
    liczby = [float(el) for el in oczyszczony_wiersz]
    dane.append(liczby);

# Definicja funkcji liczącej minima dla każdego wiersza
def znajdz_minima_wierszy(lista_danych):
    minima = []
    for wiersz in lista_danych:
        # Funkcja min() znajduje najmniejszą liczbę w danej liście (wierszu)
        najmniejsza = min(wiersz)
        minima.append(najmniejsza)
    return minima

# Wywołanie funkcji i wyświetlenie wyników
wyniki_min = znajdz_minima_wierszy(dane)

print("Wczytane dane:")
print(dane)
print("\nMinimalne wartości dla kolejnych wierszy:")
for i, wartosc in enumerate(wyniki_min, start=1):
    print(f"Wiersz {i}: {wartosc}")

print(dane)
```
