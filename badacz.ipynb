```python
with open('dane_numeryczne.csv', 'r') as plik:
  plik.readline();
  odczytane_dane = plik.readlines();
  dane = [];
  for el in odczytane_dane:
    oczyszczony_wiersz = el.strip().split(",");
    liczby = [float(el) for el in oczyszczony_wiersz]
    dane.append(liczby);

print(dane)

# Funkcja licząca średnią dla każdego wiersza
def srednia_wierszy(dane):
    srednie = []
    for wiersz in dane:
        srednia = sum(wiersz) / len(wiersz)
        srednie.append(srednia)
    return srednie

# Wywołanie funkcji i wyświetlenie wyników
srednie = srednia_wierszy(dane)
for i, s in enumerate(srednie, 1):
    print(f"Średnia dla wiersza {i}: {s:.2f}")
```
